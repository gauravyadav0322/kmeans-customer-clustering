```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('Mall_Customers.csv')

df.head()

#cluster analysis

df.info()

df.describe()

df.columns

df=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
sns.scatterplot(data=df,x='Annual Income (k$)',y='Spending Score (1-100)')
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df=ss.fit_transform(df)
df=pd.DataFrame(df)
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5)
df['Kmeans Prediction']=km.fit_predict(df)
```

```python
sns.scatterplot(data=df,x=0,y=1,hue='Kmeans Prediction',palette='Set1')
plt.legend(bbox_to_anchor=(1.05,1))

plt.show()
```

```python
#Elbow method
df=pd.read_csv('Mall_Customers.csv')
df=df[['Annual Income (k$)','Spending Score (1-100)']]
ss=StandardScaler()
df=ss.fit_transform(df)
df=pd.DataFrame(df)

wcss_score=[]
for k in range(2,21):
    print(k)
    km=KMeans(n_clusters=k)
    km.fit(df)
    wcss_score.append(km.inertia_)
```

```python
plt.plot(range(2,21),wcss_score)
```
